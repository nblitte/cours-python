## Les objets : une rapide introduction

En programmation, et particulièrement en python, les objets sont des éléments extrêmements importants.

### Ce que l'on a déjà vu mais pas sous ces termes 

Tout commence avec les classes d'objet. Une classe d'objet est une grande catégorie de valeurs qui sont régies par un ensemble de lois similaires et qui peuvent fonctionner de telle ou telle manière. vous connaissez par exemple les classes `str` et les classes `list`.

````python
print(type("Bonjour le monde"))
print(type("Au revoir le monde"))
print(type([1,25]))
````

Les phrases `Bonjour le monde` et `Au revoir le monde` sont toutes deux des chaînes de caractères, issues de la classe `str`. La classe est donc une catégorie de valeurs en programmation. Les instances de classes sont elles appelées des objets. En francais donc, `Bonjour le monde` et `Au revoir le monde` sont des objets de la classe `str`.

### Les méthodes 
Les objets peuvent posséder leurs propres fonctions. Nous en avons déjà vu :

````python
print("Bonjour le monde".replace("monde", "master TNAH"))
````

Les fonctions proprent à des classes et à leurs objets sont appelées des méthodes. Elles diffèrent des fonctions habituelles par leur syntaxe : une méthode est appelée après un point `.` et est uniquement accessible pour les classes qui la possèdent. On ne peut pas par exemple faire : 

````python
[1,2,3].replace(1, 3)
````

### Les attributs
Un objet peut aussi avoir des attributs. Les attributs sont des propriétés de ces classes qui fonctionnent comme des variables ou des clés de dictionnaires :

````python
from modules_cours.chapitre5 import objet_exemple

print(objet_exemple.attribut_exemple)
objet_exemple.attribut_exemple = 1
print(objet_exemple.attribut_exemple)
# Avec une méthode qui change des choses sans qu'on le voit
objet_exemple.change_valeur_attribut_exemple(2)
print(objet_exemple.attribut_exemple)
````

### Instancer un objet
Pour créer un objet qui ne fait pas partie des types principaux, on utilise généralement le nom de la classe avec les paramètres de base dont elle a besoin. Cela ressemble à l'utilisation d'une fonction :

````python
from modules_cours.chapitre5 import ObjetExemple

objet2 = ObjetExemple()
objet2.attribut_exemple = 8
print(objet2.attribut_exemple)
print(objet_exemple.attribut_exemple)
````

## Qu'est-ce qu'une route (et quel rapport avec les URLs ?)

Allons sur le site de l'Ecole des Chartes. Nous avons l'adresse http://www.enc-sorbonne.fr/fr/enseignements. Cette adresse se divise en http://www.enc-sorbonne.fr (le domaine) et `/fr/enseignements` (le chemin). Ce chemin (*route* en anglais) correspond pour le site à une certaine page. Quand on développe un site, on ne définit pas une url complète mais seulement le chemin. Ici, `/fr/enseignements`.

Mais il est bien sûr possible de faire plus ! Par exemple, que pouvez-vous me dire de ces deux adresses : http://www.enc-sorbonne.fr/fr/cours/initiation-xml et http://www.enc-sorbonne.fr/fr/cours/panorama-ressources-electroniques-existantes ?

Le chemin est similaire jusqu'à sa dernière partie ! On trouve souvent ce genre d'adresse par exemple avec des nombres (ce qui rend la chose encore plus claire !) : https://pleiades.stoa.org/places/109321 et https://pleiades.stoa.org/places/167717 où le chemin se découpe en `/places/[nombre]`. 

### Définir des routes à paramètres

````python
@app.route("/places/<place_id>")
def chemin_place(place_id):
    return "On est sur " + place_id
````

## Les *templates*

### Introduction

Jusque là, c'est très bien, mais afficher juste une chaîne de caractère ne fait pas de grands sites. Il va nous falloir envoyer du html. Bien sûr, on pourrait écrire cela (`python cours-flask/exemple4/app.py`):

````python
@app.route("/")
def accueil():
    nom_du_site = "Gazetteer"
    return '''
<html>
    <head>
        <title>{nom}</title>
    </head>
    <body>Bienvenue !</body>
</html>'''.format(nom=nom_du_site)
````

Mais si cela marche bien pour ce petit exemple, cela deviendra vite compliqué et illisible quand vous aurez de (très) nombreuses pages avec un HTML plus complexe qu'un `title` et un `body`. C'est pour cela que de très nombreux frameworks offrent un système de *template*. Un template est un fichier HTML, XML ou équivalent qui sera lu par votre framework et complété le cas échéant par des informations que vous fournirez.

Dans Flask, les templates sont généralement mis dans un sous-dossier `templates` du dossier principal de l'application (ici : `cours-flask/exemple4/templates/accueil.html`). Les variables sont alors incluses via des doubles accolades :

```html
<html>
    <head>
        <title>{{nom}}</title>
    </head>
    <body>Bienvenue !</body>
</html>
```

Ce fichier templates est ensuite déclaré et rendu par votre application Flask :

````python
from flask import Flask, render_template

app = Flask("Application")


@app.route("/")
def accueil():
    return render_template("accueil.html", nom="Gazetteer")
````

#### Commentaire de code 

Par rapport à l'utilisation précédente, nous avons :
- importé la fonction `render_template` depuis le module `flask`
- utilisé la fonction `render_template` qui prend:
    - en premier argument le chemin du template (à partir du sous-dossier template)
    - des arguments nommés pour le reste. Ces arguments nommés sont ensuite utilisés comme des variables dans le template

## Conditions dans les templates

Les templates vont plus loin que le simple affichage de variable, ils permettent aussi de réaliser des conditions pour vérifier qu'une chose ou une autre existe ou bien qu'une valeur particulière est affichée. Par exemple, si l'on avait potentiellement une variable `lieux` qui contiendrait le nombre de lieux enregistrés dans notre index topographique.

Nous écririons donc 

```html
...

    <body>
        <h1>Bienvenue !</h1>
        <p>Il y a {{lieux|length} enregistrés, voici le premier : {{lieux[0]}}</p>
    </body>
</html>
```

**Note** : Il n'est pas possible de faire `len(lieux)` dans un template de Flask. Nous sommes obligés de passer par l'équivalent `{{lieux|length}}`

Mais le résultat ne serait pas très convivial. Afficher potentiellement `Voici le premier : ` et du vide donnera l'impression que le site ne marche pas. Et pour cause. Nous utiliserons donc un `if` :

```html
...

    <body>
        <h1>Bienvenue !</h1>
        <p>
            Il y a {{lieux|length}} enregistrés
            {% if lieux %}
            , voici le premier : {{lieux[0]}}
            {% endif %}
    </p>
    </body>
</html>
```

Un bloc `if` est alors commencé par un `{% if condition %}` ou *conditions* est une condition python classique (on se souvient qu'une liste vide est égal à False et que donc par opposition, `lieux` sera positif si il y a un seul élément dans la liste). Le bloc `{% if condition %}` se termine nécessairement par un nouveau bloc `{% endif %}`.

Il est bien sûr possible d'aller plus loin qu'un `if` : on peut compléter ce bloc par un `{% elif %}` et/ou un `{% else %}` 

```html
...

    <body>
        <h1>Bienvenue !</h1>
        <p>
            {% if lieux %}
                Il y a {{lieux|length}} enregistrés , voici le premier : {{lieux[0]}}
            {% else %}
                La base de données est en cours de constitution
            {% endif %}
    </p>
    </body>
</html>
```

````python
from flask import Flask, render_template

app = Flask("Application")


@app.route("/")
def accueil(exemple=None):
    lieux = ["Col. Lugdunum", "Augustomagus"]
    return render_template("accueil.html", nom="Gazetteer", lieux=lieux)
````

## Boucles dans les templates et liens

Il ne manquait donc plus qu'une chose du python traditionnel: les boucles. Une fois ce type d'expression apprise, vous auraient l'ensemble des expressions importantes dans le système de template de Flask. Tout comme les `if`, les expressions de boucles se font entre des `{% %}` : `{% for valeur in conteneur %}` et `{%endfor%}`.

```html
Il y a {{lieux|length}} enregistrés :
<ul>
    {% for lieu in lieux %}
        <li>{{lieu.nom}}</li>
    {% endfor %}
</ul>
```

Le système fonctionne comme les boucles `for` en python : `{% for _ in _ %}` est donc aussi valable pour `dict.items()`, `dict.values()`, `dict.keys()`, *etc.*

À partir de cette boucle, il serait intéressant de se servir de cette page d'accueil pour faire un index. Et pour faire un index, il faut des liens. Et pour faire des liens... il faut `url_for()` dans Flask ! La fonction `url_for()` est appelée du côté du template et doit prendre comme premier argument le nom de la fonction qui vous intéresse (par exemple `lieu` est le nom de notre fonction pour la route `@app.route("/place/<int:place_id>")`. Les arguments suivants correspondent aux arguments nécessaire à la construction de l'url (ici `place_id`) : `url_for("lieu", place_id=0)` retournera la valeur `/place/0`. Dans le HTML, cela donnera : 

```html
Il y a {{lieux|length}} enregistrés :
<ul>
    {% for lieu_id, lieu in lieux.items() %}
        <li><a href="{{url_for('lieu', place_id=lieu_id)}}">{{lieu.nom}}</a></li>
    {% endfor %}
</ul>
```

ou bien pour retourner à l'accueil : 

```html
<p><a href="{{url_for('accueil')}}">Retour à l'accueil</a></p>
```

## Héritage dans les templates

La composition d'un site web est souvent répétitive : on retrouve ainsi très souvent le même en-tête, le même menu, le même pied-de-page et c'est le centre du contenu qui change.

![Composition classique d'une page web en vertical](images/decoupage.site.png)

Mais nous avons déjà deux pages. Comme d'habitude, les programmeur-se-s ont cherché à éviter la répétition de tous les éléments. Pour deux raisons :

- c'est une perte de temps de répéter les éléments, même en copiant-collant
- c'est un véritable enfer de maintenance : il faut vérifier à chaque changement que l'ensemble des pages ont été modifiées.

On va donc chercher à créer un ensemble de sous-pages qui ne contiennent que l'élément *main* tandis que l'entourage de la page sera importée depuis notre dite page. Ainsi nos pages seront constituées :
- d'un bloc principal contenant les éléments voués à se répéter
- d'un bloc secondaire contenant les informations spécifiques à une page 

![Découpage des éléments et responsabilités des blocs](images/incrustation.site.png)

Pour faire cela sous Jinja, il existe un moyen simple : la définition de block via `{% block nom_de_bloc %}` et `{% endblock %}`. On définit un block dans le fichier conteneur.html que l'on étend dans les pages via `{% extends "conteneur.html" %}` (Attention, pas de `{%endextends%}` ici !

### Avec un seul bloc d'import

#### templates/conteneur.html

```html
<html>
    <head>
        <title>Gazetteer</title>
    </head>
    <body>
        <div><h1>Bienvenue sur le Gazetteer</h1></div>
        <div>{% block corps%}{%endblock%}</div>
    </body>
</html>
```

#### templates/accueil.html

```html
{% extends "conteneur.html" %}
{% block corps %}
Il y a {{lieux|length}} enregistrés :
<ul>
    {% for lieu_id, lieu in lieux.items() %}
        <li><a href="{{url_for('lieu', place_id=lieu_id)}}">{{lieu.nom}}</a></li>
    {% endfor %}
</ul>
{% endblock %}
```

### Deux blocks

Il est aussi possible d'étendre plusieurs blocs avec un seul conteneur : il suffit de déclarer l'ensemble des blocs et de les remplir le cas échéant : 

#### templates/conteneur.html

```html
<html>
    <head>
        <title>Gazetteer {%block titre %}{%endblock%}</title>
    </head>
    <body>
        <div><h1>Bienvenue sur le Gazetteer</h1></div>
        <div>{% block corps%}{%endblock%}</div>
    </body>
</html>
```

#### templates/pages/accueil.html
*Pour plus de lisibilité, il est commun de regrouper les éléments semblables dans des sous-dossiers de templates.*

```html
{% extends "conteneur.html" %}
{% block corps %}

<ul>
    {% for lieu_id, lieu in lieux.items() %}
        <li><a href="{{url_for('lieu', place_id=lieu_id)}}">{{lieu.nom}}</a></li>
    {% endfor %}
</ul>
{% endblock %}
```

#### templates/pages/places.html

```html

{% block titre %}
    {%if lieu %}| Lieu : {{lieu.nom}} {% endif %}
{% endblock %}

{% block corps %}
    {% if lieu %}
        <h1>{{lieu.nom}}</h1>
        <h2>{{lieu.moderne}}</h2>

            <dt>Longitude</dt><dd>{{lieu.latlong[1]}}</dd>
            <dt>Latitude</dt><dd>{{lieu.latlong[0]}}</dd>
            <dt>Description</dt><dd>{{lieu.description}}</dd>
            <dt>Type</dt><dd>{{lieu.type}}</dd>

    {% else %}
        <p>La base de données est en cours de constitution</p>
    {% endif %}

{% endblock %}
```

### Les includes

En plus de bloc, il existe la possibilité non pas d'étendre mais d'inclure un ensemble d'éléments. Cela devient très pratique quand on veut multiplier les squelettes principaux ou qu'un ensemble de sous-éléments se répètent (par exemple, plusieurs pages ont une visionneuse d'image). Le découpage logique se fait alors comme suit :
- un ensemble de pages
- des conteneurs qui seront étendus par ces pages
- des sous-éléments incrustable dans les conteneurs ou les pages 

![Incrustation avancée avec des sous-éléments](images/incrustation.avancee.png)

Cette inclusion se fait via `{% include "chemin/vers/inclusion/html" %}` et ne possède pas de fin `{% endincludes %}`.

Nous pouvons donc changer `conteneur.html` et créer une page `partials/metadata.html`. Notez que si vous avez des `blocks` dans un include, le block ne sera pas complété par la page.

#### conteneur.html

```html
<html>
    <head>
        <title>Gazetteer {%block titre %}{%endblock%}</title>
        {% include "partials/metadata.html" %}
    </head>
    <body>
        <div><h1>Bienvenue sur le Gazetteer</h1></div>
        <div>{% block corps%}{%endblock%}</div>
    </body>
</html>

```


#### partials/metadata.html

```html
<meta name="description" content="Gazetteer formidable développé par l'Ecole Nationale des Chartes">
<meta name="keywords" content="Gazetteer,Chartes,Ecole,Topographie,Dictionnaire,Topologie">
<meta name="author" content="Thibault Clérice">
<meta name="viewport" content="width=device-width, initial-scale=1.0">
```

```sh
cd cours-flask/exemple9
python app.py
```

## Les assets

Pour finir ce chapitre sur les bases de flask, je propose que l'on essaye de faire quelque chose d'un peu plus joli. Parce que pour l'instant... disons que le design est discutable.

Les assets, contrairement aux templates, sont des fichiers statiques : ils ne nécessitent pas de calcul ou de remplacement. Alors pour les lier, il suffit de les placer dans un dossier `/static/` dans le dossier principal de l'application et d'y mettre le nécessaire (on a chargé pour l'exemple 10 les fichiers nécessaires à bootstrap)

Pour réaliser le lien vers un fichiers statique, on utilise aussi la fonction `url_for()`. En premier paramètre, on utilise le nom `"static"` et le paramètre nommé `filename="chemin/vers/le/fichier/depuis/le/dossier/static.gif"`. Exemple pour le fichier css de bootstrap :

```html
<link rel="stylesheet" href="{{ url_for('static', filename='css/bootstrap.css')}}">
```
